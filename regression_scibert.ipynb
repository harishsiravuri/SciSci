```python
import pandas as pd
import numpy as np
import scipy
import scipy.stats
import re
import logging

# Commenting corenlp since not available on turing
# import corenlp
import nltk

from nltk import sent_tokenize
from nltk.parse import stanford
from nltk.parse import CoreNLPParser
from nltk.tree import Tree
```

```python
parser = CoreNLPParser(url='http://localhost:9000')

# Function to generate parse trees and other information
def getTextInfo(sentence=''):
    try:
        # Using the parser defined outside the function
        global parser

        # Parse the sentence using StanfordCoreNLP
        parse_string = str(list(parser.raw_parse(sentence))[0]).replace('\n', '').replace('  ', '')

        return (pd.DataFrame({'Sentence':sentence,
                              'ParseTree':parse_string
                             }, index=[0]))

    except Exception as e:
        logging.error('An error has occured: ' + str(e))
        return (pd.DataFrame({'Sentence':'Error',
                              'ParseTree':'Error'
                             }, index=[0]))
def generate_parse_trees():
    try:
        # Read the Gutenberg_File
        raw_data = pd.read_csv('scores.csv')
        
        # Initialize parser
        # This is assuming the StanfordCoreNLP server is already running
        parser = CoreNLPParser(url='http://localhost:9000')

        # Get necessary information for both AP and Gutenberg datasets
        #raw_data.apply(lambda x: compute_yules_i_for_text(x['abstract']), axis=1)
        raw_data['parse_tree'] = raw_data.apply(lambda x: getTextInfo(x['abstract']), axis=1)
        return(raw_data)
    
    except Exception as e:
        logging.error('An error has occured: ' + str(e))
        return(pd.DataFrame())
```
